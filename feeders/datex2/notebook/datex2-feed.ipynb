```python
EVENTSTREAM_NAME = 'datex2-ingest'
STATE_FILE = '/lakehouse/default/Files/feeder-state/datex2/state.json'
POLLING_INTERVAL = 300
ONCE_MODE = True
WORKSPACE_ID = ''
DATEX2_SOURCES = ""  # optional source selector: comma-separated catalog entry names, "*" for all, empty for the catalog's enabled defaults
```

```python
import os, sys, threading
from notebookutils import notebook
os.environ['STATE_FILE'] = STATE_FILE
if DATEX2_SOURCES:
    os.environ['DATEX2_SOURCES'] = DATEX2_SOURCES
os.environ['ONCE_MODE'] = 'true' if ONCE_MODE else 'false'
from datex2_kafka import app as feeder
sys.argv = ['datex2_kafka', 'feed', '--once']
errors = []
def run():
    try:
        feeder.main()
    except Exception as exc:
        errors.append(exc)
t = threading.Thread(target=run)
t.start(); t.join()
if errors:
    notebook.exit('FAIL: ' + str(errors[0]))
notebook.exit('OK')
```
